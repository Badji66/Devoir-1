```python
import random

# 1. Créer une liste de nombres de 1 à 10 et afficher chaque nombre multiplié par 2
nombres = list(range(1, 11))
print("Multiplication par 2:", [n * 2 for n in nombres])

# 2. Créer une liste de noms et afficher la longueur de chaque nom
noms = ["Mouhamed", "Astou", "Babacar", "daouda"]
print("Longueur des noms:", {nom: len(nom) for nom in noms})

# 3. Fonction qui prend une liste et renvoie une nouvelle liste contenant le carré de chaque élément
def carre_liste(liste):
    return [x ** 2 for x in liste]

print("Carré des nombres:", carre_liste(nombres))

# 4. Créer une liste de nombres aléatoires et afficher seulement ceux supérieurs à 5
nombres_aleatoires = [random.randint(1, 10) for _ in range(10)]
nombres_sup_5 = [x for x in nombres_aleatoires if x > 5]
print("Nombres aléatoires:", nombres_aleatoires)
print("Nombres > 5:", nombres_sup_5)

# 5. Programme qui prend une liste de chaînes de caractères et affiche la plus longue
def chaine_la_plus_longue(chaines):
    return max(chaines, key=len) if chaines else "Liste vide"

chaines = ["pomme", "banane", "cerise", "ananas"]
print("Chaîne la plus longue:", chaine_la_plus_longue(chaines))
```

```python
def elements_communs(liste1, liste2):
    """Renvoie une liste des éléments communs entre liste1 et liste2"""
    return list(set(liste1) & set(liste2))

def pair_ou_impair(nombre):
    """Renvoie 'pair' si le nombre est pair, sinon 'impair'"""
    return "pair" if nombre % 2 == 0 else "impair"

def somme_liste(liste):
    """Renvoie la somme des éléments de la liste"""
    return sum(liste)

def premieres_lettres(mots):
    """Renvoie une liste contenant la première lettre de chaque mot"""
    return [mot[0] for mot in mots if mot]

def supprimer_voyelles(chaine):
    """Renvoie la chaîne sans voyelles"""
    voyelles = "aeiouAEIOU"
    return "".join(lettre for lettre in chaine if lettre not in voyelles)

# Exemples d'utilisation :
print(elements_communs([1, 2, 3, 4], [3, 4, 5, 6]))  # [3, 4]
print(pair_ou_impair(7))  # "impair"
print(somme_liste([1, 2, 3, 4]))  # 10
print(premieres_lettres(["chat", "chien", "oiseau"]))  # ['c', 'c', 'o']
print(supprimer_voyelles("Bonjour tout le monde"))  # "Bnjr tt l mnd"
```

```python
import statistics

def mediane(liste):
    return statistics.median(liste)

liste_nombres = [3, 1, 4, 1, 5, 9, 2, 6, 5]
print("Médiane:", mediane(liste_nombres))


def filtrer_longues_chaines(liste):
    return [chaine for chaine in liste if len(chaine) > 5]

liste_chaines = ["chat", "éléphant", "pomme", "ordinateur", "soleil"]
print("Chaînes de plus de 5 caractères:", filtrer_longues_chaines(liste_chaines))


def moyenne_sans_sum(liste):
    total = 0
    for nombre in liste:
        total += nombre
    return total / len(liste) if liste else 0

liste_nombres = [10, 20, 30, 40, 50]
print("Moyenne:", moyenne_sans_sum(liste_nombres))


def inverser_liste(liste):
    return liste[::-1]

liste_exemple = [1, 2, 3, 4, 5]
print("Liste inversée:", inverser_liste(liste_exemple))


liste_noms = ["Alice", "Bob", "Charlie", "David", "Eve"]
noms_avec_a = [nom for nom in liste_noms if "a" in nom.lower()]
print("Noms contenant 'a':", noms_avec_a)
```

```python
def remove_duplicates(lst):
    """Renvoie une liste sans les éléments en doublon."""
    return list(set(lst))


def generate_primes(n):
    """Génère une liste de nombres premiers jusqu'à un nombre donné."""
    primes = []
    for num in range(2, n + 1):
        is_prime = all(num % p != 0 for p in primes)
        if is_prime:
            primes.append(num)
    return primes


def multiply_by_three(lst):
    """Multiplie chaque élément d'une liste par 3 en utilisant map."""
    return list(map(lambda x: x * 3, lst))


# Exemples d'utilisation
test_list = [1, 2, 2, 3, 4, 4, 5]
print("Liste sans doublons :", remove_duplicates(test_list))

max_number = 20
print("Nombres premiers jusqu'à", max_number, ":", generate_primes(max_number))

numbers = [1, 2, 3, 4, 5]
print("Multiplication par 3 :", multiply_by_three(numbers))
```

```python
from collections import Counter
import string

def letter_frequency(text: str) -> dict:
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Filtrer uniquement les lettres de l'alphabet
    letters = [char for char in text if char in string.ascii_lowercase or char in 'àâäéèêëîïôöùûüç']
    
    # Compter les occurrences de chaque lettre
    letter_counts = Counter(letters)
    
    # Calculer la fréquence relative
    total_letters = sum(letter_counts.values())
    frequency = {letter: count / total_letters for letter, count in letter_counts.items()}
    
    return frequency

# Exemple d'utilisation
text = "Bonjour, ceci est un exemple en français et en English!"
print(letter_frequency(text))
```
